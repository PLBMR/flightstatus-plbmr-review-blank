```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pdpbox import pdp
import joblib
```

```python
# load the model
grid_XGB_class = joblib.load('grid_XGB_class')
m_xgb = grid_XGB_class.best_estimator_
```

```python
# load the data
flights = pd.read_csv('/Users/user/Documents/Github/flightstatus/data/dev/flights_processed.csv')
X_data = flights.drop(['TARGET'], axis=1)
```

```python
def get_pdp(feature):
    pdp_goals = pdp.pdp_isolate(model=m_xgb, dataset=X_data, model_features=X_data.columns, feature=feature)
    pdp.pdp_plot(pdp_goals, feature)
    plt.show()
```
